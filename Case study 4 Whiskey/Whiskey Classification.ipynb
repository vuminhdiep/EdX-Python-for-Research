```python
import numpy as np
import pandas as pd
whiskey = pd.read_csv("whiskies.txt")
print(type(whiskey))
whiskey["Region"] = pd.read_csv("regions.txt")
whiskey.head()
```

```python
flavors = whiskey.iloc[:,2:14]
flavors
```

```python
corr_flavors = pd.DataFrame.corr(flavors)
print(corr_flavors)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.savefig("correlation_flavors.pdf")
```

```python
corr_whiskey = pd.DataFrame.corr(flavors.transpose())
plt.figure(figsize=(10,10))
plt.pcolor(corr_whiskey)
plt.axis("tight")
plt.colorbar()
plt.savefig("correlation_whiskey.pdf")
```

```python
from sklearn.cluster.bicluster import SpectralCoclustering
model = SpectralCoclustering(n_clusters=6,random_state=0)
model.fit(corr_whiskey)
model.rows_
np.sum(model.rows_,axis=1) #output tells how many whiskey belong to cluster 1, 2, ..
np.sum(model.rows_,axis=0) #output tells how many clusters belong to each observation
```

```python
model.row_labels_ #output means observation 0 belongs to cluster 2, observation 1 belong to cluster 4, observation 2 belongs to cluster 5
```

```python
whiskey['Group'] = pd.Series(model.row_labels_,index=whiskey.index)
whiskey = whiskey.loc[np.argsort(model.row_labels_)]
whiskey = whiskey.reset_index(drop=True)
```

```python
correlations = pd.DataFrame.corr(whiskey.iloc[:,2:14].transpose())
correlations = np.array(correlations)
```

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.pcolor(corr_whiskey)
plt.title('Original')
plt.axis('tight')
plt.subplot(122)
plt.pcolor(correlations)
plt.title('Rearranged')
plt.axis('tight')
plt.savefig('correlations.pdf')
```
